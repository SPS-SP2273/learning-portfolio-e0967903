## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
```

```python
x = [0,1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]


plt.plot(x, y_1, c='red', ls='dashed', label='Y values', lw=2)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y',c='red')
plt.grid(alpha=.25) #translucency of bg grid 
plt.legend(loc='upper left')
plt.show() # will show output even without this but good to include so that the code will work in non-interactive environments
```

### 1.2 Adding another plot

```python
x = [0,1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

plt.plot(x, y_1, c='red', ls='dashed', label='Y values', lw=2)
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25) 
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='best')
# supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()

#can the line smoothness be improved?
# ans: np.linspace()
```

### 1.7 We need another axis!

```python
#x = np.array([0, 1, 2, 3, 4, 5, 6])
x=np.linspace(0,6,100)
# 0 is the start. 6 is the end. 100 is the number of samples to generate (makes line smoother with bigger number)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx() # new y axis 
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')
# there are two plt.legend(), one for each line plotted


plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')


plt.savefig('simple-plot.png', dpi=150)
# saves in the same folder as this notebook (plotting)
# path needs to be specified in greater detail to save the file elsewhere
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
plt.style.use('default')
# even if this line is deleted, the style remains
# plt.style.use('default') to reset it to default style

#plt.xkcd()
# the font family not found errors are rather irritating especially since the effect still works

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.7)
plt.hlines(0,date[0],date[-1],ls='--',colors='white')
# 0 is y value, date[0] is the first year, date[-1] is the last year

plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
# the values look different from the website...?
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.linspace(0,50,500)
y = np.e**(-x/10)*np.sin(x)

plt.plot(x,y)
plt.ylim=[-1,1]
plt.yticks(np.arange(-1.00, 1.25, 0.25))
plt.xlabel('X',fontsize=15,c='darkblue')
plt.ylabel('Y',fontsize=15,c='darkblue',rotation=0)
plt.title('$y=e^{-x/10}sin(x)$',fontsize=20)
plt.grid(alpha=.25)
plt.show()



```
